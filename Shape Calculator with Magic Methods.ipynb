Shape Calculator with Magic Methods

```python
#1.
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

#2.
    def __str__(self):
        return f"Rectangle(width={self.width}, height={self.height})"
    
#3.
    def area(self):
        return self.width * self.height

#4.
    def __add__(self, other):
        return Rectangle(self.width + other.width, self.height + other.height)

#5.
    def __eq__(self, other):
        return self.area() == other.area()

#6.
rect1 = Rectangle(4, 5)
rect2 = Rectangle(2, 3)

#7.
print(f"Rectangle 1: {rect1}")
print(f"Rectangle 2: {rect2}")

print(f"\nArea of rect1: {rect1.area()}")
print(f"Area of rect2: {rect2.area()}")

#8.
rect3 = rect1 + rect2
print(f"\nrect3 (rect1 + rect2): {rect3}")


rect4 = Rectangle(6, 1)  
print(f"rect4             : {rect4}")


print(f"\nAre rect1 and rect2 equal? {rect1 == rect2}")
print(f"Are rect2 and rect4 equal? {rect2 == rect4}")
```
