Write a script that processes a range of numbers using loops and conditional statements.

```python
# Part 1: FizzBuzz using for loop
for num in range(1, 21):

    if num % 3 == 0 and num % 5 == 0:
        print("FizzBuzz")

    elif num % 3 == 0:
        print("Fizz")

    elif num % 5 == 0:
        print("Buzz")

    else:
        print(num)

# Part 2: Countdown using while loop
num = int(input("Enter a number for countdown: "))

while num > 0:

    if num > 10:
        print("Number is greater than 10. Stopping countdown.")
        break

    print(num)
    num -= 1

print("Blast Off!")
```
