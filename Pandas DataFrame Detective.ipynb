Pandas DataFrame Detective

```python
import pandas as pd
import numpy as np

# Creating the dictionary
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve', 'Frank'],
    'Age': [25, 30, np.nan, 35, 28, np.nan],
    'Score': [85, 92, 78, np.nan, 88, 95],
    'Grade': ['A', 'A', 'B', 'C', 'B', 'A']
}

# Create DataFrame
df = pd.DataFrame(data)

#Perform Inspections

# df.head(3) - Returns the first 3 rows
print("First 3 Rows")
print(df.head(3))

# df.shape - Returns (rows, columns)
print("\nDataFrame Shape")
print(df.shape)

# df.columns - Returns column names
print("\nDataFrame Columns")
print(df.columns)

# df.dtypes - Returns data types of each column
print("\nData Types")
print(df.dtypes)

# df.isnull() - Checks for missing values (returns Booleans)
print("\nIs Null?")
print(df.isnull())

# Fill Age with mean
# In newer pandas versions, it's best to specify numeric_only=True or fill a specific column
mean_age = df['Age'].mean()
df['Age'] = df['Age'].fillna(mean_age)
print(f"\nMean age used for filling: {mean_age:.1f}")

# Drop rows with missing Score
df = df.dropna(subset=['Score'])

print("\nDataFrame after handling missing data")
print(df)


# Select Name and Score
print("Name and Score Columns")
print(df[['Name', 'Score']])

# Select rows where Score > 85
print("\nScores Greater Than 85")
high_scores = df[df['Score'] > 85]
print(high_scores)

# Set Name as index
df = df.set_index('Name')
print("\nDataFrame with Name as Index")
print(df)


# Unique grades
print("Unique Grades")
print(df['Grade'].unique())

# Grade counts
print("\nGrade Counts")
print(df['Grade'].value_counts())

# Sort by Score descending
print("\nSorted by Score (Descending)")
df_sorted = df.sort_values(by='Score', ascending=False)
print(df_sorted)
```
