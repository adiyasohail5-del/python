1. Variables (Name, Age, City)

```python
name = "Ali"
age = 20
city = "Gujranwala"

print(name, age, city)
```

2. Data Types

```python
num_int = 10
num_float = 3.14
text = "Hello"
is_student = True

print(type(num_int))
print(type(num_float))
print(type(text))
print(type(is_student))
```

3. Arithmetic Operations

```python
a = 10
b = 5

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
```

4. Comparison Operators

```python
x = 8
y = 12

print(x > y)   # greater than
print(x < y)   # less than
print(x == y)  # equal to
```

5. Type Conversion

```python
num = 10
num_float = float(num)
print(num_float, type(num_float))

num2 = 5.9
num_int = int(num2)
print(num_int, type(num_int))
```

6. Student Information

```python
student_name = "Ali"
roll_number = 101
marks = 85

print("Name:", student_name)
print("Roll Number:", roll_number)
print("Marks:", marks)
```

7. Assignment Operators

```python
value = 10

value += 5
print(value)

value -= 3
print(value)

value *= 2
print(value)
```

8. Boolean Operators

```python
a = True
b = False

print(a and b)
print(a or b)
print(not a)
```

9. Area of Rectangle

```python
length = 10
width = 5

area = length * width
print("Area of rectangle:", area)
```
