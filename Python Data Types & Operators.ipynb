Python Data Types & Operators

```python
# 1. Basic Data Types
age = 20
cgpa = 3.4
name = "Ali"
is_student = True
print("Data Types:")
print(type(age), type(cgpa), type(name), type(is_student))


# 2. List (Multiple Students)
marks = [85, 78, 92, 60, 45]
print("\nMarks List:", marks)

# 3. Tuple (Subjects - Fixed)
subjects = ("Math", "Physics", "CS")
print("Subjects:", subjects)

# 4. Dictionary (Student Info)
student = {
    "name": name,
    "age": age,
    "cgpa": cgpa,
    "marks": marks
}

print("\nStudent Info:", student)

# 5. Arithmetic Operators
total_marks = sum(marks)
average_marks = total_marks / len(marks)

print("\nTotal Marks:", total_marks)
print("Average Marks:", average_marks)

# 6. Comparison Operators

print("\nIs average > 50?", average_marks > 50)
print("Is average == 60?", average_marks == 60)

# 7. Logical Operators

if average_marks > 50 and cgpa > 2.5:
    print("Student is doing well")
else:
    print("Student needs improvement")



# 8. Assignment Operators

bonus = 5
average_marks += bonus 

print("\nAverage after bonus:", average_marks)

# 9. Membership Operators

print("\nIs 92 in marks?", 92 in marks)
print("Is 30 not in marks?", 30 not in marks)



# 10. Decision Making 

if average_marks >= 50:
    result = "Pass"
else:
    result = "Fail"

print("\nFinal Result:", result)

# 11. Loop (Extra Skill)

print("\nIndividual Marks:")
for m in marks:
    print(m)

# 12. Mini Problem Solving

# Find highest and lowest marks
highest = max(marks)
lowest = min(marks)

print("\nHighest Marks:", highest)
print("Lowest Marks:", lowest)

# 13. 

print("\n--- Final Report ---")
print("Name:", student["name"])
print("Age:", student["age"])
print("CGPA:", student["cgpa"])
print("Average Marks:", average_marks)
print("Result:", result)
```
