ASSIGNMENT 1: The Data Analyzer

```python
import math  # Square root ke liye

# Step 1: Function to get numbers from user
def get_numbers():
    numbers = input("Enter numbers separated by spaces: ")
    # Convert each input to float
    data = [float(num) for num in numbers.split()]
    return data

# Step 2: Function to calculate mean
def calculate_mean(data):
    return sum(data) / len(data)

# Step 3: Function to calculate sample variance
def calculate_variance(data, mean):
    n = len(data)
    variance = sum((x - mean) ** 2 for x in data) / (n - 1)  # sample variance
    return variance

# Step 4: Function to calculate standard deviation
def calculate_std_dev(variance):
    return math.sqrt(variance)

# Main program
data = get_numbers()
mean = calculate_mean(data)
variance = calculate_variance(data, mean)
std_dev = calculate_std_dev(variance)

# Step 5: Print results
print("\n--- Statistical Analysis ---")
print("Data:", data)
print("Mean:", round(mean, 2))
print("Variance:", round(variance, 2))
print("Standard Deviation:", round(std_dev, 2))
```

ASSIGNMENT 2: The Student Class Syste

```python
# Base class Person
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_info(self):
        print("\n--- Person ---")
        print("Name:", self.name)
        print("Age:", self.age)

# Derived class Student
class Student(Person):
    def __init__(self, name, age, student_id, major):
        super().__init__(name, age)  # Call base class constructor
        self.student_id = student_id
        self.major = major

    def display_info(self):
        print("\n--- Student ---")
        print("Name:", self.name)
        print("Age:", self.age)
        print("Student ID:", self.student_id)
        print("Major:", self.major)

# Derived class GraduateStudent
class GraduateStudent(Student):
    def __init__(self, name, age, student_id, major, thesis_topic):
        super().__init__(name, age, student_id, major)  # Call Student constructor
        self.thesis_topic = thesis_topic

    def display_info(self):
        print("\n--- Graduate Student ---")
        print("Name:", self.name)
        print("Age:", self.age)
        print("Student ID:", self.student_id)
        print("Major:", self.major)
        print("Thesis Topic:", self.thesis_topic)

# Create objects
person1 = Person("John Doe", 45)
student1 = Student("Jane Smith", 20, "S12345", "Computer Science")
grad_student1 = GraduateStudent("Bob Johnson", 27, "G67890", "Physics", "Quantum Computing")

# Call display_info()
person1.display_info()
student1.display_info()
grad_student1.display_info()
```

ASSIGNMENT 3: Shape Calculator with Magic Methods

```python
# Step 1: Create Rectangle class
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    # Step 2: __str__ method for readable output
    def __str__(self):
        return f"Rectangle(width={self.width}, height={self.height})"

    # Step 3: area() method
    def area(self):
        return self.width * self.height

    # Step 4: Overload + operator
    def __add__(self, other):
        # Sum widths and heights to create new Rectangle
        new_width = self.width + other.width
        new_height = self.height + other.height
        return Rectangle(new_width, new_height)

    # Step 5: Overload == operator to compare areas
    def __eq__(self, other):
        return self.area() == other.area()


# Step 6: Create rectangles
rect1 = Rectangle(4, 5)
rect2 = Rectangle(2, 3)

# Step 7: Print rectangles and areas
print("Rectangle 1:", rect1)
print("Rectangle 2:", rect2)
print("Area of rect1:", rect1.area())
print("Area of rect2:", rect2.area())

# Step 8: Create rect3 by adding rect1 and rect2
rect3 = rect1 + rect2
print("\nrect3 (rect1 + rect2):", rect3)
print("Area of rect3:", rect3.area())

# Step 9: Compare rectangles
print("\nAre rect1 and rect2 equal?", rect1 == rect2)

# Extra check: Compare rect2 and a rectangle with same area
rect4 = Rectangle(1, 6)
print("Are rect2 and rect4 equal?", rect2 == rect4)
```
