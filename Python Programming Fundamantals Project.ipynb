```python
def calculate_average(scores: list) -> float:
    """Calculates and returns the average of a list of numbers."""
    if not scores:
        return 0.0
    return sum(scores) / len(scores)


def determine_letter_grade(average_score: float) -> str:
    """Takes a numerical score and returns the corresponding letter grade."""
    if average_score >= 90:
        return 'A'
    elif average_score >= 80:
        return 'B'
    elif average_score >= 70:
        return 'C'
    elif average_score >= 60:
        return 'D'
    else:
        return 'F'
# CORE FEATURES
def add_student(database: dict):
    """Handles the user input to add a new student and their scores."""
    print("\nAdd New Student")
    name = input("Enter student name: ").strip().capitalize()
    
    if not name:
        print("❌ Name cannot be empty!")
        return
        
    # Get scores from user
    scores = []
    print("Enter scores one by one. Type 'done' when finished.")
    while True:
        score_input = input("Enter a score (0-100): ").strip().lower()
        if score_input == 'done':
            break
            
        try:
            score = float(score_input)
            if 0 <= score <= 100:
                scores.append(score)
            else:
                print("❌ Score must be between 0 and 100.")
        except ValueError:
            print("❌ Invalid input. Please enter a number or 'done'.")
            
    if not scores:
        print("❌ No scores entered. Student not added.")
        return
        
    # Store in dictionary
    database[name] = scores
    print(f"✅ Successfully added {name}!")


def display_report(database: dict):
    """Prints a structured summary of all students and their performance."""
    if not database:
        print("\nℹ️ No student data available yet.")
        return
        
    print("\n" + "=" * 45)
    print(f"{'STUDENT REPORT':^45}")
    print("=" * 45)
    print(f"{'Name':<12} {'Scores':<20} {'Avg':<8} {'Grade':<5}")
    print("-" * 45)
    
    for name, scores in database.items():
        avg = calculate_average(scores)
        grade = determine_letter_grade(avg)
        
        # Format scores list as a clean string
        scores_str = ", ".join(map(str, [int(s) if s.is_integer() else s for s in scores]))
        
        print(f"{name:<12} {scores_str:<20} {avg:<8.1f} {grade:<5}")
        
    print("=" * 45)
# MAIN PROGRAM LOOP
def main():
    # Basic data structure to hold our information
    student_database = {
        'Alice': [85.0, 92.0, 88.0],
        'Bob': [70.0, 65.0, 72.0]
    }
    
    while True:
        print("\nMAIN MENU")
        print("1. Add Student")
        print("2. Display Report")
        print("3. Exit")
        
        choice = input("Select an option (1-3): ").strip()
        
        if choice == '1':
            add_student(student_database)
        elif choice == '2':
            display_report(student_database)
        elif choice == '3':
            print("\nThank you for using the Grade Manager. Goodbye!")
            break
        else:
            print("❌ Invalid choice! Please select 1, 2, or 3.")

# This ensures the program runs when executed directly
if __name__ == '__main__':
    main()
```
