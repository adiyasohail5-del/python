```python
try:
    # Step 1: input numbers
    num1 = float(input("Enter first number: "))
    operator = input("Enter operator (+, -, *, /): ")
    num2 = float(input("Enter second number: "))

    # Step 2: conditions
    if operator == "+":
        result = num1 + num2
        print("Result:", result)

    elif operator == "-":
        result = num1 - num2
        print("Result:", result)

    elif operator == "*":
        result = num1 * num2
        print("Result:", result)

    elif operator == "/":
        if num2 != 0:
            result = num1 / num2
            print("Result:", result)
        else:
            print("Invalid input! Division by zero not allowed.")

    else:
        print("Invalid input! Operator not supported.")

except ValueError:
    print("Invalid input! Please enter numeric values only.")
```
