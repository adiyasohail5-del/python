Loops & Functions

```python
import random

# 1. VARIABLES & DATA TYPES
player_name = "Ali"      # String
score = 0                # Integer
high_score = 0.0         # Float
is_playing = True        # Boolean

print("Welcome to Number Games!")
print("Player:", player_name)

# 2. FUNCTION - Even or Odd
def even_or_odd(number):
    if number % 2 == 0:
        return "Even"
    else:
        return "Odd"

# Calling the function
num = 7
result = even_or_odd(num)
print(num, "is", result)

# 3. FOR LOOP - FizzBuzz
print("\nFizzBuzz from 1 to 20:")

for i in range(1, 21):
    if i % 3 == 0 and i % 5 == 0:
        print("FizzBuzz")
    elif i % 3 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz")
    else:
        print(i)
# 4. WHILE LOOP - Guess the Number
print("\nGuess the Number Game!")
secret = random.randint(1, 10)
attempts = 0
while attempts < 5:
    guess = int(input("Guess (1-10): "))
    attempts = attempts + 1

    if guess == secret:
        print("Correct! You won!")
        score = score + 10
        break
    elif guess < secret:
        print("Too low!")
    else:
        print("Too high!")

if guess != secret:
    print("Game Over! Number was:", secret)

print("Your Score:", score)
# 5. OPERATORS
print("\nCalculator:")

a = 20
b = 4

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Modulus:", a % b)
```
