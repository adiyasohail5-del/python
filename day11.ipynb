```python
class Student:
    def __init__(self):
        self.__name = "ali"      # private variable
        self.age = 20
        self.salary = 0         # salary variable

    def display(self):
        print("Name:", self.__name)
        print("Age:", self.age)
        print("Salary:", self.salary)

    def get_name(self):
        return self.__name

    def set_name(self, name):
        self.__name = name

    def set_salary(self, salary):
        self.salary = salary   # fixed

    def calculate_bonus(self):
        if self.salary >= 5000:
            bonus = self.salary * 0.05   # 5% bonus
        else:
            bonus = 0
            print("No bonus")
        return bonus


# object created
s1 = Student()

# user se salary input
sal = float(input("Enter your salary: "))
s1.set_salary(sal)

print("\nStudent Details:")
s1.display()

bonus = s1.calculate_bonus()
print("5% Bonus:", bonus)
```

```python
   # Parent Class
class Shape:
    def __init__(self, color):
        self.color = color

    def display(self):
        print("This is a", self.color, "shape")


# Child Class 1
class Circle(Shape):
    def __init__(self, radius):
        super().__init__("Circle")
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius


# Child Class 2
class Square(Shape):
    def __init__(self, side):
        super().__init__("Square")
        self.side = side

    def area(self):
        return self.side * self.side


# Child Class 3
class Rectangle(Shape):
    def __init__(self, length, width):
        super().__init__("Rectangle")
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width
    
#Child class 4
class Triangle(Shape):
    def __init__(self, base, height):
        super().__init__("Triangle")
        self.base = base
        self.height = height

    def area(self):
        return 0.5 * self.base * self.height
    
circle_radius = float(input("Enter radius of the circle: "))
square_side = float(input("Enter side length of the square: "))
rectangle_length = float(input("Enter length of the rectangle: "))
rectangle_width = float(input("Enter width of the rectangle: "))
triangle_base = float(input("Enter base of the triangle: "))
triangle_height = float(input("Enter height of the triangle: "))

# Creating Objects
c = Circle(circle_radius)
s = Square(square_side)
r = Rectangle(rectangle_length, rectangle_width)
t = Triangle(triangle_base, triangle_height)

# Calling Methods
c.display()
print("Area:", c.area())

s.display()
print("Area:", s.area())

r.display()
print("Area:", r.area())

t.display()
print("Area:", t.area())
```

```python
# Parent Class
class Device:
    def __init__(self, brand):
        self.brand = brand

    def power_on(self):
        print(f"{self.brand} device is powering on...")

# Child Class
class Laptop(Device):
    def __init__(self, brand, ram):
        super().__init__(brand)   # Call parent constructor
        self.ram = ram

    def show_specs(self):
        print(f"Brand: {self.brand}")
        print(f"RAM: {self.ram} GB")

l1 = input("Enter laptop brand: ")
l2 = int(input("Enter RAM size (GB): "))
# Example Usage
l1 = Laptop(l1, l2)

l1.power_on()
l1.show_specs()
    
        
```

```python
# Parent Class
class Employee:
    def __init__(self, name, employee_id, salary):
        self.__name = name                # Private attribute
        self.__employee_id = employee_id  # Private attribute
        self.__salary = salary            # Private attribute
        self._department = "General"      # Protected attribute

    # Getter Methods
    def get_name(self):
        return self.__name

    def get_employee_id(self):
        return self.__employee_id

    def get_salary(self):
        return self.__salary

    # Setter Method
    def set_salary(self, new_salary):
        if new_salary > 0:
            self.__salary = new_salary
        else:
            print("Salary must be positive.")

    # Bonus Calculation
    def calculate_bonus(self):
        return self.__salary * 0.05  # 5% bonus

    # Display Info
    def display_info(self):
        print("Name:", self.__name)
        print("Employee ID:", self.__employee_id)
        print("Salary:", self.__salary)
        print("Department:", self._department)


# Child Class
class Manager(Employee):
    def __init__(self, name, employee_id, salary, team_size):
        super().__init__(name, employee_id, salary)
        self.__team_size = team_size      # Private attribute
        self._department = "Management"  # Modify protected attribute

    # Getter for team size
    def get_team_size(self):
        return self.__team_size

    # Method Overriding
    def calculate_bonus(self):
        # Managers get 10% bonus instead of 5%
        return self.get_salary() * 0.10

    # Overriding display method
    def display_info(self):
        super().display_info()
        print("Team Size:", self.__team_size)

e1 = input("Enter employee name: ")
e2 = input("Enter employee ID: ")
e3 = int(input("Enter employee salary: "))

m1 = input("Enter manager name: ")
m2 = (input("Enter manager ID: "))
m3 = int(input("Enter manager salary: "))
m4 = int(input("Enter manager team size: "))
# Object Creation
e1 = Employee(e1, e2, e3)
m1 = Manager(m1, m2, m3, m4)

print("---- Employee Info ----")
e1.display_info()
print("Bonus:", e1.calculate_bonus())

print("\n---- Manager Info ----")
m1.display_info()
print("Bonus:", m1.calculate_bonus())
```

```python
class Car:
    def drive(self):
        print("Driving a car")

class SportsCar(Car):
    def drive(self):
        print("Driving a sports car at high speed")

class Truck(Car):
    def drive(self):
        print("Driving a truck with heavy load")

class ElectricCar(Car):
    def drive(self):
        print("Driving an electric car silently")


# Polymorphism in action
cars = [Car(), SportsCar(), Truck(), ElectricCar()]

for vehicle in cars:
    vehicle.drive()   # Same method name, different behavior
```

```python
class Car:
    def __init__(self, name, brand):
        self.name = name
        self.brand = brand

    def drive(self):
        print(f"Driving a {self.brand} {self.name}")


# Child Classes
class SportsCar(Car):
    def drive(self):
        print(f"Driving a sports car {self.brand} {self.name} at high speed")


class Truck(Car):
    def drive(self):
        print(f"Driving a {self.brand} {self.name} with heavy load")


class ElectricCar(Car):
    def drive(self):
        print(f"Driving an electric {self.brand} {self.name} silently")

car = input("Enter car name: ")
brand = input("Enter car brand: ")

SportsCar = SportsCar(car, brand)
Truck = Truck(car, brand)
ElectricCar = ElectricCar(car, brand)

cars = [SportsCar, Truck, ElectricCar]

for vehicle in cars:
    vehicle.drive()
```

```python
class vehical:
    def __init__(self, brand, model, year):
        self.brand = brand
        self.model = model
        self.year = year

    def display_info(self):
        print(f"Brand: {self.brand}")
        print(f"Model: {self.model}")
        print(f"Year: {self.year}")

class Car(vehical):
    def __init__(self, brand, model, year, num_doors):
        super().__init__(brand, model, year)
        self.num_doors = num_doors

    def display_info(self):
        super().display_info()
        print(f"Number of Doors: {self.num_doors}")

class Motorcycle(vehical):
    def __init__(self, brand, model, year, has_sidecar)
        super().__init__(brand, model, year)
        self.has_sidecar = has_sidecar

    def display_info(self):
        super().display_info()
        print(f"Has Sidecar: {'Yes' if self.has_sidecar else 'No'}")

car_brand = input("Enter car brand: ")
car_model = input("Enter car model: ")
car_year = input("Enter car year: ")
car_doors = input("Enter number of doors: ")

motorcycle_brand = input("Enter motorcycle brand: ")
motorcycle_model = input("Enter motorcycle model: ")
motorcycle_year = input("Enter motorcycle year: ")
motorcycle_sidecar = input("Does motorcycle have sidecar? (yes/no): ").lower() == "yes"

car = Car(car_brand, car_model, car_year, car_doors)
motorcycle = Motorcycle(motorcycle_brand, motorcycle_model, motorcycle_year, motorcycle_sidecar)

# Display information for each vehicle
print("\nCar Information:")
car.display_info()

print("\nMotorcycle Information:")
motorcycle.display_info()
```

```python
class vehicale:
    def __init__(self, brand):
        self.brand = brand
    def start(self):
        print(f"{self.brand} is starting...")

class Car(vehicale):
    def __init__(self, brand, model):
        super().__init__(brand)
        self.model = model

    def display_info(self):
        print(f"Brand: {self.brand}")
        print(f"Model: {self.model}")

class Bike(vehicale):
    def __init__(self, brand, type):
        super().__init__(brand)
        self.type = type

    def display_info(self):
        print(f"Brand: {self.brand}")
        print(f"Type: {self.type}")

car_brand = input("Enter car brand: ")
car_model = input("Enter car model: ")
bike_brand = input("Enter bike brand: ")
bike_type = input("Enter bike type: ")

car = Car(car_brand, car_model)
bike = Bike(bike_brand, bike_type)

print("\nCar Information:")
car.display_info()
print("\nBike Information:")
bike.display_info()
```
