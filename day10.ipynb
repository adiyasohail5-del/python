```python
class BankAccount:
    def __init__(self):
        self.account_balance = float(input("Enter account balance: "))
    def deposit (self):
        amount = float (input("Enter deposite amount:"))
        self.account_balance += amount
        print("Account balance:", self.account_balance)
    def withdraw(self):
        withdraw = float(input("Enter withdrawal amount: "))
        if withdraw <= self.account_balance:
            self.account_balance -= withdraw
            print("Account balance:", self.account_balance)
        else:

            print("Insufficient funds")
    def show_balance(self):
        print("your account balance is:", self.account_balance)

obj = BankAccount()
obj.deposit()
obj.withdraw()
obj.show_balance()
```

```python
class employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
        self.increase_amount = 0
    def increase_salary(self, percent):
       if percent > 0:
        self.increase_amount += self.salary * (percent / 100)
        self.salary += self.increase_amount
        print(f"salary increased by {percent}%")
       else:
        print("Invalid percentage. Please enter a positive value.")

    def display(self):
        print("Employee Name:", self.name)
        print("Employee Salary:", self.salary)
    
emp1 = employee("Ali", 50000)
emp1.display()

emp1.increase_salary(2)
emp1.display()
```

```python
class calculator:
    def __init__(self):
        self.num1 = int(input("Enter first number: "))
        self.num2 = int(input("Enter second number: "))
    def add(self):
        return self.num1 + self.num2
    def subtract(self):
        return self.num1 - self.num2
    def multiply(self):
        return self.num1 * self.num2
    def divide(self):
        if self.num2 != 0:
            return self.num1 / self.num2
        else:
            return "Error: Division by zero is not allowed."
obj = calculator()
print("Addition:", obj.add())
print("Subtraction:", obj.subtract())
print("Multiplication:", obj.multiply())
print("Division:", obj.divide())
```

```python
class number:
    def __init__(self):
        self.num = int(input("Enter a number: "))
    def is_even(self):
        return self.num % 2 == 0
    def is_odd(self):
        return self.num % 2 != 0
    
obj = number()
print("Is the number even?", obj.is_even())
print("Is the number odd?", obj.is_odd())
```

```python
class Movies:
    def __init__(self):
        self.movie_list = []

        n = int(input("How many movies do you want to enter? "))

        for i in range(n):
            movie = input("Enter movie name: ")
            self.movie_list.append(movie)

    def show_movies(self):
        print("\nMovie List:")
        for movie in self.movie_list:
            print(movie)


# Creating object
m1 = Movies()

# Showing movies
m1.show_movies()
```

```python
class WordCounter:
    def __init__(self, sentence):
        self.sentence = sentence
        self.word_counts = {}  # dictionary to store counts

    def count_words(self):
        words = self.sentence.split()  # split sentence into words
        for word in words:
            word = word.lower()  # optional: make all words lowercase
            if word in self.word_counts:
                self.word_counts[word] += 1
            else:
                self.word_counts[word] = 1

    def display(self):
        print("Word Counts:")
        for word, count in self.word_counts.items():
            print(f"{word}: {count}")


# Example usage
sentence = "Apple mango apple banana apple"
counter = WordCounter(sentence)
counter.count_words()
counter.display()
```
