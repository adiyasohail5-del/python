```python
pd.read_csv(r"c:\Users\BM\Downloads\weather.csv")
df = pd.read_csv(r"c:\Users\BM\Downloads\weather.csv")
print(df.describe())
print(df.info())
print(df.isnull().sum().sum())
print(df.duplicated())
print(df.drop_duplicates())
print(df.iloc[2])
print(df.loc[0:2])
print(df.value_counts())
print(sorted(df['Time'].unique()))
print(df.merge(df, on='Description', how='inner'))
df["MAX temp (deg Celcius)"] = df["Max temp (deg Celsius)"].astype(int)
print(df.groupby("Location")["Max temp (deg Celsius)"].mean())
print(df.ffill())
```
