Assignment 1

```python
# Personal Assistant Program

# Step 1: Ask user's name
name = input("What is your name? ")

# Step 2: Ask birth year
birth_year = int(input("What year were you born? "))

# Step 3: Ask current age
age = int(input("What is your current age? "))

# Step 4: Calculate year when user turns 100
current_year = 2026
year_turn_100 = 100 - age + current_year

# Step 5: Ask lucky number
lucky_number = float(input("Enter your lucky number: "))

# Step 6: Print summary
print("\n--- Personal Assistant Summary ---")
print("Hello,", name)
print("You were born in", birth_year)
print("You are currently", age, "years old")
print("You will turn 100 in the year", year_turn_100)
print("Your lucky number is", lucky_number)
```

Assignment 2

```python
# Shopping List Manager

# Step 1: Create shopping list
shopping_list = ["milk", "bread", "eggs", "apples", "chicken"]

# Step 2: Print list and third item
print("Initial Shopping List:", shopping_list)
print("The third item is:", shopping_list[2])

# Step 3: Add two new items
shopping_list.append("cheese")
shopping_list.append("coffee")

# Step 4: Remove one item
shopping_list.remove('chicken')
print(shopping_list)

# Step 5: Create dictionary of item prices
item_prices = {
    "milk": 2.5,
    "bread": 1.99,
    "eggs": 3.25,
    "apples": 0.99
}

# Step 6: Print price of specific item
print("The price of bread is $", item_prices["bread"])

# Step 7: Print final summary
print("Updated Shopping List:", shopping_list)
print("Item Prices:", item_prices)

print("\n--- Final Summary ---")
print("Items to buy:", ", ".join(shopping_list))
print("Price lookup:", item_prices)
```

```python
# Assignment 3: Number Cruncher

# Part 1: FizzBuzz using for loop
for num in range(1, 21):

    if num % 3 == 0 and num % 5 == 0:
        print("FizzBuzz")

    elif num % 3 == 0:
        print("Fizz")

    elif num % 5 == 0:
        print("Buzz")

    else:
        print(num)

# Part 2: Countdown using while loop
num = int(input("Enter a number for countdown: "))

while num > 0:

    if num > 10:
        print("Number is greater than 10. Stopping countdown.")
        break

    print(num)
    num -= 1

print("Blast Off!")
```
